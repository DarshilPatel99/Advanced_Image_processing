```python
#useful modules

import cv2
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
i = cv2.cvtColor(cv2.imread("forest.jpg"),cv2.COLOR_BGR2RGB)/255

plt.imshow(i)
plt.rcParams["figure.figsize"]=10,10
plt.show()
```

```python
T = cv2.cvtColor(cv2.imread("trans.jpg"),cv2.COLOR_BGR2RGB)/255
plt.imshow(T)
plt.rcParams["figure.figsize"]=10,10
plt.show()
```

```python
A = [0.8159, 0.8186, 0.8272]
r,c,v  = i.shape
A_r = np.full((r,c), A[0])
A_g = np.full((r,c), A[1])
A_b = np.full((r,c), A[2])
A = cv2.merge((A_r, A_g, A_b))
```

```python
j = np.array(np.divide(i-np.multiply((1-T),A),T)*255,dtype='uint8')
```

```python
plt.imshow(j)
plt.rcParams["figure.figsize"]=10,10
plt.show()
```
